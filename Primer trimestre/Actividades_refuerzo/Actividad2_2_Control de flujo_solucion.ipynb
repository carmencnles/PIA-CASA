**1.**- Escribe un código que pida por teclado una tempratura Celsius o Farenheit con el siguiente formato:
23C, 67F...
Si introduce una C la final (grados celsius) debes devolver la temperatura en Farenheit. Si introduce una F, lo contrario. Si el texto introducido no corresponde con el formato debes indicarlo.
Se ha incluido la primera línea para pedir el texto por teclado.

```python
temperatura = input("Introduzca la temperatura (Por ejemplo: 47F, 100C...) : ")

if len(temperatura) < 2:
    print("El valor introducido es demasiado corto")
else:
    try:
    
        grados = int(temperatura[:-1])
        unidades = temperatura[-1].upper()

        if unidades == "C":
          resultado = int(round((9 * grados) / 5 + 32))
          unidades = "Fahrenheit"
          print(f"La temperatura en {unidades} es {resultado} grados.")

        elif unidades == "F":
          resultado = int(round((grados - 32) * 5 / 9))
          unidades = "Celsius"
          print(f"La temperatura en {unidades} es {resultado} grados.")
          
        else:
          print("El valor introducido está mal.")

    except ValueError:
      print("El valor introducido está mal. El formato debe ser un número seguido de una letra (ej: 45C).")
```

**3.**- Haz un programa que genere una lista de números y compruebe y muestre por pantalla cuantos hay pares y cuantos impares.


```python
lista_numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]
pares,impares = 0,0
for numero in lista_numeros:
        if numero % 2 == 0:
                pares = pares + 1
        else:
                impares = impares + 1
print("El número de pares es " + str(pares) + " y el de impares " + str(impares))
```
