**2.**-  Escribe un programa que pida nombre de usuario y contraseña. Debes hacerlo con un while que lo repita un máximo de 10 veces o hasta que lo meta bien. El usuario debe ser "Neo" y la contraseña "followTWR99".

```python
usuario = ""
contrasena = ""
intentos = 1
#El próximo while comprueba el usuario y contraseña así como el número de intentos.
while not (usuario == "Neo" and contrasena == "followTWR99") and intentos <= 10:
    print("Es su intento número " + str(intentos))
    usuario = input("Introduzca su nombre de usuario: ")
    contrasena = input("Introduzca su contraseña: ")
    intentos+=1
#Si no ha salido del while por el número de intentos te dice que has conseguido acceder.
#En caso contrario quiere decir que has hecho 10 intentos y te lo muestra.
if intentos <= 10:
    print("Acceso Permitido")
else:
    print("Ha alcanzado el máximo número de intentos.")

```

**3.**- 
Escribe un programa que pida números enteros de uno en uno, los meta en una lista y, cuando acabe (debes pedir introducir una letra en concreto para acabar) debe recorrer la lista y mostrar la suma de los pares y la suma de los impares.

```python
import re
lista_numeros = []
elemento = ""
numeros_pares = []
numeros_impares = []
print("Debes introducir los números de una lista. Introduce una q para salir.")
#Hasta que no introduzca una q sigue pidiendo números. En el momento en que sea q, sale del while.
while elemento != "q":
    elemento = input("Introduce un nuevo elemento: ")
    #Comprueba que es un número:
    if elemento.isnumeric():
        #lo cambia a entero ya que todo lo que cogemos por teclado es string:
        elemento = int(elemento)
        #lo introduce en la lista:
        lista_numeros.append(elemento)
    #recorremos la lista comprobando si son números pares o impares:
    for numero in lista_numeros:
        if numero%2==0:
            numeros_pares.append(numero)
        else:
            numeros_impares.append(numero)
print("La suma de los números pares es: " + str(sum(numeros_pares)))
print("La suma de los números impares es: " + str(sum(numeros_impares)))
```

1) crea una lista nueva.

2) añade elemento a la lista.

4) muestra solo los elementos pares.

5) salir.
