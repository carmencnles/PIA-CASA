**1.**-  Escribe in programa que pida un número al usuario y muestre su tabla de multiplicar. El número debe ser del 1 al 9. Debes hacerlo con una sentencia while.

```python
# escribe tu código
numero = int(input('Introduce un número del 0 al 9 '))

i = 1

while i <= 10:
    resultado = numero * i
    print(f"{numero} * {i} = {resultado}")
    i = i + 1

```

**2.**-  Escribe un programa que pida nombre de usuario y contraseña. Debes hacerlo con un while que lo repita un máximo de 10 veces o hasta que lo meta bien. El usuario debe ser "Neo" y la contraseña "followTWR99".

```python
# Escribe código

i = 1

while i <= 10:
    usuario = input('Introduce el nombre de usuario')
    contraseña = input('Introduce una contraseña correcta')

    if usuario == "Neo" and contraseña == "followTWR99":
        break
    else:
        print("Usuario incorrecto y contraseña incorrectos")
        i = i + 1

#Con este codigo lo pide maximo 10 veces, pero si queremos pintar y avisar que se han agotado los intentos:
if i > 10:
    print("Has agotado los intentos.")
```

**3.**- 
Escribe un programa que pida números enteros de uno en uno, los meta en una lista y, cuando acabe (debes pedir introducir una letra en concreto para acabar) debe recorrer la lista y mostrar la suma de los pares y la suma de los impares.

```python
# escribe el código

print("Introduce números enteros. Escribe 'F' (o cualquier letra) para terminar.")

lista = []

while True:
    entrada = input("Dato: ")
    if entrada.upper() == "F":
        break
    try:
        numero = int(entrada)
        lista.append(numero)
    except ValueError:
        print("Esto no es un numero valido. Introduce un numero o 'F' para salir")

sum_pares = 0
sum_impares = 0

for num in lista:
    if num % 2 == 0:
        sum_pares = sum_pares + num
    else:
        sum_impares = sum_impares + num

print(lista)
print(f"La suma de los pares es: {sum_pares}")
print(f"La suma de los pares es: {sum_impares}")






```

**4.-** Escribe un programa que muestre un menú (debes hacerlo con un while) como este:

1) crea una lista nueva.

2) añade elemento a la lista.

3) muestra los elementos de tipo numérico.

4) muestra solo los elementos pares.

5) salir.


En la primera opción debe crear una lista nueva. Es decir, si ya se había creado una y se habían añadido elementos, se crea una nueva vacía y se pierde.
En la segunda opción (si el usuario pulsa un 2) debe solicitar un nuevo elemento y lo mete en la lista. Si el elemento es un número debe meterlo como tal, sino, como string.
En la tercera opción el programa debe recorrer la lista y mostrar sólo los elementos de tipo numérico.
En la cuarta solo los números que sean pares.
Si el usuario introduce un 5 el programa debe terminar.

```python
# escribe el código

continuar = True
list = []
while continuar:
    print("\n--- MENÚ PRINCIPAL ---")
    print("1) Crea una lista nueva (borra la anterior)")
    print("2) Añade elemento a la lista")
    print("3) Muestra los elementos de tipo numérico")
    print("4) Muestra solo los elementos pares")
    print("5) Salir")

    try:
        menu = int(input("Seleccione el número de la opcion que desea: "))
    except ValueError:
        print("Error, por favor introduce un número valido.")
        continue

    if menu == 1:
        list = []
        print("Se ha creado una lista nueva")

    elif menu == 2:
        elemento = input("Introduce un nuevo elemento")
        try:
            valor = int(elemento)
            list.append(valor)
            print(f"Añadido el número: {valor}")
        except ValueError:
            list.append(elemento)
            print(f"Añadido el texto: {elemento}")
        list.append(elemento)

    elif menu == 3:
        for item in list:
            if isinstance(item, int):
                print(item)

    elif menu == 4:
        print("Numeros pares en la lista: ")
        for item in list:
            if isinstance(item, int) and item % 2 == 0:
                print(item)


    elif menu == 5:
        continuar = False

    else:
        print("Opción no válida, elija del 1 al 5")
```
