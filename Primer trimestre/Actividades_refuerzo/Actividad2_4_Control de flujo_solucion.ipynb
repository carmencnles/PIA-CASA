**1.**-  Crea una lista con datos numéricos aleatorios entre 0 y 99. Haz un menú usando while que muestre las siguientes opciones:

1. Añade elemento a la lista.
2. Elimina elemento de la lista.
3. Muestra la media de los datos.
4. Muestra la mediana de los datos.
5. Muestra el elemento máximo (debes recorrer la lista y comparar los elementos en vez de usar una función o método).
6. Salir.

```python
# SOLUCIÓN
import random
import numpy as np
lista = []
for i in range(1,random.randint(2,50)):
    lista.append(random.randint(0,99))
while True:
    print(lista)
    #print(len(lista))
    print("""Elige una opción de las siguientes:
    1. Añade elemento a la lista.
    2. Elimina elemento de la lista.
    3. Muestra la media de los datos.
    4. Muestra la mediana de los datos.
    5. Muestra el elemento máximo (debes recorrer la lista y comparar los elementos en vez de usar una función o método).
    6. Salir.""")
    opcion = input("Tu opción es: ")
    if opcion == "1":
        elemento = input("Dame un elemento: ")
        if elemento.isnumeric():
            lista.append(int(elemento))
        else:
            print("No has metido un número.")
    elif opcion == "2":
        posicion = input("Introduce la posición del elemento que quieres borrar.")
        if posicion.isnumeric() and int(posicion) in range(0,len(lista)):
            lista.pop(int(posicion))
    elif opcion == "3":
        print("La media es " + str(np.mean(lista)))
    elif opcion == "4":
        print("La mediana es " + str(np.median(lista)))
    elif opcion == "5":
        maximo = 0
        for numero in lista:
            if numero>maximo:
                maximo=numero
        print("El valor máximo es " + str(maximo))
    elif opcion == "6":
        break
    else:
        print("La opción elegida es incorrecta.")
```
