**1.**-  Crea una lista con datos numéricos aleatorios entre 0 y 99. Haz un menú usando while que muestre las siguientes opciones:

1. Añade elemento a la lista.
2. Elimina elemento de la lista.
3. Muestra la media de los datos.
4. Muestra la mediana de los datos.
5. Muestra el elemento máximo (debes recorrer la lista y comparar los elementos en vez de usar una función o método).
6. Salir.

```python
import random
import numpy as np


#aleatorios = [random.randint(0, 100)]
list = []
continuar = True

for i in range(1, random.randint(0, 100)):
    list.append(random.randint(0, 100))

while continuar:
    print("\n--- MENÚ PRINCIPAL ---")
    print("1) Añade elemento a la lista")
    print("2) Elimina elemento a la lista")
    print("3) Muestra la media de los datos")
    print("4) Muestra la mediana a los datos")
    print("5) Muestra el elemento máximo")
    print("6) Salir")

    
    try:
        menu = int(input("Seleccione el número de la opcion que desea: "))
    except ValueError:
        print("Error, por favor introduce un número valido.")
        continue

    if menu == 1:
        elemento = int(input("Introduce un nuevo elemento"))
        list.append(elemento)
        print("Elemento añadido")
        print("Lista completa: ", list)

    elif menu == 2:
        elemento = int(input("Introduce un elemento a eliminar "))
        list.remove(elemento)

    elif menu == 3:
        print("La media es: " + str(np.mean(list)))

    elif menu == 4:
        print("La mediana: " + str(np.median(list)))

    elif menu == 5:
        var = 0
        for item in list:
            if item > var:
                var = item
        print(var)
    elif menu == 6:
        continuar = False

    else:
        print("Opción no válida, elija del 1 al 6")


```
