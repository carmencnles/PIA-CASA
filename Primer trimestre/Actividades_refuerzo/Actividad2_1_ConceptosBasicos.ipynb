**1.**- Escribe una lista con datos heterogéneos. 

1.1 Haz un programa que intercambie el primer elemento por el último.

```python
# escribe el código


def intercambia1 (Lista, pos_primera, pos_ultima):
    Lista[pos_primera], Lista[pos_ultima] = Lista[pos_ultima], Lista[pos_primera]
    return Lista

Lista = ["hola", 2, "que tal", [1,2,3], 4]

print(intercambia1(Lista, pos_primera=0, pos_ultima=4))


#Otra manera de hacerlo:

def intercambiar_primero_ultimo(Lista):
    # Verificamos que la lista tenga al menos 2 elementos - Opcional
    if len(Lista) >= 2:
        # Intercambiamos el índice 0 (primero) con el -1 (último)
        Lista[0], Lista[-1] = Lista[-1], Lista[0]
    return Lista

Lista = ["hola", 2, "que tal", [1,2,3], 4]

print(intercambiar_primero_ultimo(Lista))
```

1.2 Haz una función que intercambie dos elementos de una lista y recibiendo como parámetros la lista y las posiciones.

```python
# Función intercambia
def Intercambia(List, pos1, pos2):
    List[pos1], List[pos2] = List[pos2], List[pos1]     

    return List
 
# Probamos la función
List = [10, 12, 23, 71]
pos1, pos2  = 1, 3
print(Intercambia(List,pos1,pos2))
```

**3.**- Busca todas los métodos que tienen los diccionarios en Python.

3.1 Haz un código probando el uso de cada uno de ellos.


```python
# escribe el código

diccionario = {"Nombre": "Carmen", "Edad": 22, "Año de nacimiento": 2003, "Lenguaje utilizado": "Python"}

diccionario.clear() # Elimina todos los elementos.
diccionario.copy() # Devuelve una copia poco profunda del diccionario.
diccionario.fromkeys(['N', 'E', 'A', 'L', '****']) # Crea un nuevo diccionario con las claves de la secuencia seq
# y los valores establecidos por value. El valor el opcional y por defecto es none.
diccionario.get(['Nombre']) #Devuelve el valor correspondiente a la clave key.
diccionario.items() # Devuelve una vista de los pares clave-valor del diccionario,
# que puede ser convertida en una lista de tuplas.
diccionario.keys() # Devuelve una vista de claves del diccionario, que puede ser convertida en una lista.
diccionario.pop(["Nombre"]) # Elimina el elemento con la clave key y devuelve su valor. Si la clave no existe
# y se añade default, devuelve default, si no da un keyerror.
diccionario.popitem() # Elimina y devuelve un par clave-valor aleatorio del diccionario.
diccionario.setdefault(["Hola"]) # Si la clave existe, devuelve su valor. Si no existe inserta la clave con el valor default.
diccionario.update() # Actualiza el diccionario con los apres clave-valor del iterable. Si la clave existe se reemplaza, si no, se añade.
diccionario.values() # Devuelve una vista de los valores del diccionario que puede ser convertida en una lista.

```
